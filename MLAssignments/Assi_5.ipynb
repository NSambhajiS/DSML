Q] Visualize the data using R/Python by plotting the graphs for assignment no. 1 and 2. Consider
suitable data set.  
a) Use Scatter plot, bar plot, Box plot and Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('car_crashes')
print(df.head)
```

```python
#scatter plot
plt.scatter(df.speeding,df.alcohol)
plt.show()
```

```python
#bar plot
sns.barplot(x='alcohol',y='speeding',data=df)
plt.show()
```

```python
#box plot
sns.boxplot(x="alcohol",data=df)
```

```python
#histogram
df.hist(column='alcohol' )
plt.show()
```
